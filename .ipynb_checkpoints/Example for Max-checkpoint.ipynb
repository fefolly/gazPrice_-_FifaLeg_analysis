### Load Libraries

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### Real World Examples

### Download data from github

```python
gas = pd.read_csv('gas_prices.csv')

#gas  #to display the data
#print(gas.Year)

#Alter size of the graph
plt.figure(figsize=(8,5))

plt.title('Gas Prices over Time (in USD)', fontdict={'fontweight':'bold', 'fontsize': 22})

#plot graph
plt.plot(gas.Year, gas.USA,  'b.-', label='USA')
plt.plot(gas.Year, gas.Canada, label='Canada')
plt.plot(gas.Year, gas['South Korea'], label='South Korean')


plt.xticks(gas.Year[::3]) #skiping every 3 year in the graph

#add label
plt.xlabel('Year')
plt.ylabel('US Dollars')

plt.legend() #to show the legends

#save the graph
#plt.savefig('Gas_price_figure.pdf', dpi=300)
#plt.savefig('Gas_price_figure.png', dpi=300)

plt.show()
```

### Load Fifa Data

```python
fifa = pd.read_csv('fifa_data.csv')

fifa
```

### Histograms

```python
#setting an interval
bins = [40,50,60,70,80,90,100]

plt.figure(figsize=(8,5))
plt.hist(fifa.Overall, bins=bins)


#set the x axis internal as the bins
plt.xticks(bins)

plt.ylabel('Number of Players')
plt.xlabel('Skill Level')
plt.title('Distribution of player Skills in FIFA 2018', fontdict={'fontweight':'bold', 'fontsize': 18})

plt.show()
```

### Pie Charts

```python
#fifa['Preferred Foot']
```

```python
left = fifa.loc[fifa['Preferred Foot'] == 'Left'].count()[0]
right = fifa.loc[fifa['Preferred Foot'] == 'Right'].count()[0]


labels = ['Left', 'Right']
colors = ['#abcdef', '#aabbcc']

plt.pie([left, right], labels = labels, colors=colors, autopct='%.2f %%')

plt.title('Foot Preference of FIFA Players', fontdict={'fontweight':'bold', 'fontsize': 18})

plt.show()
```

### Weight

```python
fifa.Weight =  [int(x.strip('lbs')) if type(x)==str else x for x in fifa.Weight]

#fifa.Weight
plt.style.use('ggplot')
plt.figure(figsize=(15,15))

light = fifa.loc[fifa.Weight < 125].count()[0]
light_mediun = fifa.loc[(fifa.Weight >= 125) & (fifa.Weight < 150)].count()[0]
medium = fifa.loc[(fifa.Weight >= 150) & (fifa.Weight < 175)].count()[0]
medium_heavy = fifa.loc[(fifa.Weight >= 175) & (fifa.Weight < 200)].count()[0]
heavy =  fifa.loc[(fifa.Weight > 200)].count()[0]


weights = [light, light_mediun, medium, medium_heavy, heavy]
labels = ['Under 125', '125-150', '150-175', '175-200', 'Over 200']

explode = (.4,.2,.0,.0,.4)

plt.title('Weight Distribution of FIFA Players (in lbs)')

plt.pie(weights, labels=labels, autopct='%.2f %%', pctdistance=0.8, explode=explode)

plt.show()
```

```python
plt.style.use('default')

plt.figure(figsize=(7,8))

barcelona = fifa.loc[fifa.Club == 'FC Barcelona']['Overall']
madrid = fifa.loc[fifa.Club == 'Real Madrid']['Overall']
revs = fifa.loc[fifa.Club  == 'New England Revolution']['Overall']


labels = ['FC Barcelona', 'Real Madrid', 'New England Revolution']

plt.boxplot([barcelona, madrid, revs], labels=labels)

plt.title('Professional Soccer Team Comparision')
plt.ylabel('FIFA Overall Rating')



plt.show()
```
